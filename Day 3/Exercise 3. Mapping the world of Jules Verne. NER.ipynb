```python
#Solution
with open("ving_mille_day3.txt", "r", encoding = "utf-8") as f:
    data = f.read()
```

#### B. We import the libraries

```python
#Your code in here
```

```python
#Solution
import spacy
```

#### C. We download the French pipeline

```python
#Your code in here
```

```python
#Solution
nlp = spacy.load("fr_core_news_sm")
```

#### D. We initialize the NLP object

```python
#Your code in here
```

```python
#Solution
doc = nlp(data)
```

#### E. We create 1 list with LOC (probably ORG has some relevant results, but for reasons of practicality, let's focus on LOC!). Let's also remove the entity labels and create a set to drop dupplicated values.

```python
#Your code in here
```

```python
#Solution
loc = []

for ent in doc.ents:
    if ent.label_ == "LOC":
        loc.append([ent.text, ent.label_])
```

Now let's remove the label

```python
location_places = []

for a, b in loc:
    location_places.append(a)
```

And now let's get rid of dupplicated values

```python
unique_loc = set(location_places)
```

```python
unique_loc
```

#### F. Let's save our values

```python
#Your code in here
```

```python
with open("LOC_vingtmille.txt", "w", encoding = "utf-8") as f:
    f.write(str(unique_loc))
```
